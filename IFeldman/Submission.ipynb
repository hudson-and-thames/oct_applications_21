# Implementation of Kinectic Component Analysis

## Introduction
This notebook demonstrates an implementation of the Kinectic Component Analysis (KCA) from the paper: Kinectic Component Analsyis by Marcos Lopez de Prado and Ricardo Rebonato.

The KCA algorithm extracts a signal from a series of noisy measurements by applying the Kalman Filter on Taylor expansion of a stochastic process.  The algorithm decomposes the signal into three hidden components: position, velocity, acceleration.  This can be useful to the analysis of trend-following, momentum and mean-reverting financial prices.

## 3. Conclusion

The current implimentation and use of the KCAMomentum and KCAMeanReversion alpha strategies are naive implementations and are subject to the problems of backtesting errors addressed in several papers by MLdP, David Bailey and others.

As for much of my thoughts on FinML - it is not advisable to use ML algorithms without thorough consideration of the Alpha Modeling questions you are trying to answer.

- What questions are you trying to asnwer?
- Is it answerable?
- Is it valuable?
- Is the ML technique / algorithm applicable to this types of questions?

In the case of the KCA algorithm we have an algorithm that decomposes a financial time series into 3 hidden factors: position, velocity and acceleration.  The natural place to use this is with a time series that has a strong mean reversion tendency.  From that it follows that this ML algorithm could be used for pairs trading.  Taking to the next step would be trading a basket of securities - like in this case the currency pairs of USD, EUR, JPY and GBP - and use the sensitivity of the KCA factors to determine when a currency pair is showing momentum or mean reversion tendencies.
